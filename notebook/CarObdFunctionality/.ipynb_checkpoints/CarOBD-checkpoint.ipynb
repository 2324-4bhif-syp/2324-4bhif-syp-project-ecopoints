```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei laden
df = pd.read_csv("input/obd_functionality.csv", encoding='ISO-8859-1', sep=';')

# Funktionalitätsdaten ausschließen, um nur die Autos zu haben
df_autos = df.drop(columns=["functionality"])

# Zähle die Werte (0, 1, 2) für jedes Auto
count_0 = (df_autos == 0).sum()
count_1 = (df_autos == 1).sum()
count_2 = (df_autos == 2).sum()

# Zusammenführen der Zähler in ein DataFrame
counts = pd.DataFrame({
    'Funktioniert (0)': count_0,
    'Funktioniert nicht (1)': count_1,
    'Liest falsch aus (2)': count_2
})

# Gestapeltes Balkendiagramm erstellen
ax = counts.plot(kind='bar', stacked=True, color=['#2ca02c', '#d62728', '#ff7f0e'], figsize=(10, 6))

# Diagramm anpassen
plt.title('OBD-Funktionalität bei verschiedenen Fahrzeugen')
plt.xlabel('Fahrzeuge')
plt.ylabel('Anzahl der Funktionalitäten')
plt.xticks(rotation=45)
plt.legend(title='Status der Funktionalität')
plt.tight_layout()

plt.yticks(range(0, 21, 1))

# Diagramm anzeigen
plt.show()
```

```python
# Transponiere die Tabelle, damit die OBD-Funktionen zu Zeilen und die Fahrzeuge zu Spalten werden
df_transposed = df.set_index("functionality").T

# Zähle, wie oft jede Funktion (Zeile) funktioniert (0 und 2 gelten als funktionierend)
working_count = (df_transposed == 0).sum() + (df_transposed == 2).sum()

# Sortiere die Funktionalitäten nach der Anzahl der Fahrzeuge, bei denen sie funktionieren
working_count_sorted = working_count.sort_values(ascending=False)

# Barplot erstellen, um die Funktionalitäten darzustellen, die am meisten funktionieren
plt.figure(figsize=(10, 6))
working_count_sorted.plot(kind='bar', color='#2ca02c')

# Diagramm anpassen
plt.title('OBD-Funktionen, die am meisten funktionieren')
plt.xlabel('Funktionalitäten')
plt.ylabel('Anzahl der Fahrzeuge, bei denen sie funktionieren')
plt.xticks(rotation=90)
plt.tight_layout()

# Diagramm anzeigen
plt.show()
```
