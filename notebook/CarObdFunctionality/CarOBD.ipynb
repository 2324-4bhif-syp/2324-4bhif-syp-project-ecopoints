```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei laden
df = pd.read_csv("input/obd_functionality.csv", encoding='ISO-8859-1', sep=';')

# Funktionalittsdaten ausschlieen, um nur die Autos zu haben
df_autos = df.drop(columns=["functionality"])

# Zhle die Werte (0, 1, 2) fr jedes Auto
count_0 = (df_autos == 0).sum()
count_1 = (df_autos == 1).sum()
count_2 = (df_autos == 2).sum()

# Zusammenfhren der Zhler in ein DataFrame
counts = pd.DataFrame({
    'Funktioniert (0)': count_0,
    'Funktioniert nicht (1)': count_1,
    'Liest falsch aus (2)': count_2
})

# Gestapeltes Balkendiagramm erstellen
ax = counts.plot(kind='bar', stacked=True, color=['#2ca02c', '#d62728', '#ff7f0e'], figsize=(10, 6))

# Diagramm anpassen
plt.title('OBD-Funktionalitt bei verschiedenen Fahrzeugen')
plt.xlabel('Fahrzeuge')
plt.ylabel('Anzahl der Funktionalitten')
plt.xticks(rotation=45)
plt.legend(title='Status der Funktionalitt')
plt.tight_layout()

plt.yticks(range(0, 21, 1))

# Diagramm anzeigen
plt.show()
```

```python
# Transponiere die Tabelle, damit die OBD-Funktionen zu Zeilen und die Fahrzeuge zu Spalten werden
df_transposed = df.set_index("functionality").T

# Zhle, wie oft jede Funktion funktioniert (0) und wie oft sie falsch ausgelesen wird (2)
working_count = (df_transposed == 0).sum()  # Funktioniert (0)
faulty_count = (df_transposed == 2).sum()   # Falsch ausgelesen (2)

# Zusammenfhren der Zhler in ein DataFrame
counts = pd.DataFrame({
    'Funktioniert (0)': working_count,
    'Liest falsch aus (2)': faulty_count
})

# Gestapeltes Balkendiagramm erstellen
ax = counts.plot(kind='bar', stacked=True, color=['#2ca02c', '#ff7f0e'], figsize=(10, 6))

# Diagramm anpassen
plt.title('OBD-Funktionen und ihre Funktionsfhigkeit')
plt.xlabel('Funktionalitten')
plt.ylabel('Anzahl der Fahrzeuge')
plt.xticks(rotation=90)
plt.legend(title='Status der Funktionalitt')
plt.tight_layout()

# Diagramm anzeigen
plt.show()
```
